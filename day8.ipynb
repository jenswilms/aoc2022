```python
file = open("./input8.txt", "r").read().split('\n')[:-1]

def isVisible(treemap, y, x):
    # Edges
    if y == 0 or y == len(treemap) - 1 or x == 0 or x == len(treemap[y]) - 1:
        return True
    
    # Get height of tree at (y, x)
    height = int(treemap[y][x])
    
    # Up
    higherTreeExist = False
    for y1 in range(y - 1, -1, -1):
        if int(treemap[y1][x]) >= height:
            higherTreeExist = True
            break        
    if higherTreeExist == False:
        return True    
     
    # Down
    higherTreeExist = False
    for y1 in range(y + 1, len(treemap), 1):
        if int(treemap[y1][x]) >= height:
            higherTreeExist = True
            break    
    if higherTreeExist == False:
        return True    
    
    # Left
    higherTreeExist = False
    for x1 in range(x - 1, -1, -1):
        if int(treemap[y][x1]) >= height:
            higherTreeExist = True
            break
    if higherTreeExist == False:
        return True    
    
    # Right
    higherTreeExist = False
    for tree in treemap[y][x+1:]:
        if int(tree) >= height:
    # for x1 in range(x + 1, len(treemap[y]), 1):
    #     if int(treemap[y][x1]) >= height:
            higherTreeExist = True
            break     
    if higherTreeExist == False:
        return True 
    
    # If no taller trees found, return True
    return False

def checkTrees(treemap):
    # Initialize empty list of visible trees
    visibleTrees = []
    
    # Iterate over rows and columns of the treemap
    for index_y, row in enumerate(treemap):
        for index_x, number in enumerate(row):
            # Check if tree at (index_y, index_x) is visible
            if isVisible(treemap, index_y, index_x):
                # If visible, add coordinates to visibleTrees list
                visibleTrees.append((index_y, index_x))
    
    # Return number of visible trees
    return len(visibleTrees)

print(checkTrees(file))
```
