Options for programming:
* draw out total points for all 9 scenarios and then calculate (easiest but ugliest)
* transform letters into categories, then compare, and then add points for the shape choice
* transform letters into numerical variables, then compare, add points for shape choice
* create conditions for winning

```python
input = open("./input2.txt", "r")
totalpoints = 0
for newline in input:
    line = newline[0:3]
    if line == "A X": 
        totalpoints += 3 + 1
    if line == "A Y":
        totalpoints += 6 + 2
    if line == "A Z":
        totalpoints += 0 + 3
    if line == "B X":
        totalpoints += 0 + 1
    if line == "B Y":
        totalpoints += 3 + 2
    if line == "B Z":
        totalpoints += 6 + 3
    if line == "C X":
        totalpoints += 6 + 1
    if line == "C Y":
        totalpoints += 0 + 2
    if line == "C Z":
        totalpoints += 3 + 3

print(totalpoints)
```

```python
input = open("./input2.txt", "r")
def transformSymbol(symbol):
    if symbol == "A" or symbol == "X":
        return 1
    
    if symbol == "B" or symbol == "Y":
        return 2
    
    if symbol == "C" or symbol == "Z":
        return 3
    
def playRound(opponent, yourvalue):
    # lose (default)
    points = 0
    # draw
    if opponent == yourvalue:
        points = 3
    
    # win
    if (opponent == 1 and yourvalue == 2) or (opponent == 2 and yourvalue == 3) or (opponent == 3 and yourvalue == 1):
        points = 6

    return points

totalpoints = 0
for line in input:
    symbols = line.split()
    values = []
    for symbol in symbols:
        values.append(transformSymbol(symbol))
    gamepoints = playRound(values[0], values[1])
    totalpoints += (gamepoints + values[1])
      
print(totalpoints)
```
